#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
# your code here
fig = plt.figure(1, figsize=(12, 6))
ax = fig.add_subplot(111)
boxplot = ax.boxplot(titanic['Age'])

ax.set_title('Age Distribution on Titanic')
ax.set_ylabel('Age')
ax.set_xlabel('box')
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
x = titanic['Age']

fig, [ax1, ax2, ax3] = plt.subplots(3,1)
plt.rcParams["figure.figsize"] = (6,6)
ax1.hist(x, bins=10)
ax1.set_title('Passenger Ages')
ax1.set_xlabel('Age') 
ax1.set_ylabel('Passengers')      

ax2.hist(x, bins=20)
ax2.set_title('Passenger Ages')
ax2.set_xlabel('Age')
ax2.set_ylabel('Passengers')

ax3.hist(x, bins=50)
ax3.set_title('Passenger Ages')
ax3.set_xlabel('Age')
ax3.set_ylabel('Passengers')

plt.show()

```

```python
# Whether the bin size affects your plot? Why?
#It would seem having a smaller bin size gives a better idea as to the various ages. Larger bins make it harder
#to distinguish the difference between a 10 year old and 20 year old. It makes it harder for instance, to determine
#how many adults vs kids died since we mark adulthood at 18.
```
